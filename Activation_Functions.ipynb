Task: Visualizing Different Activation Functions used in Neural Network

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_function(x,y,title):
    #plt.figure(figsize=(5,3)).gca()
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
```

### **Rectified Linear Unit (ReLU) Function**

```python
def relu(x):
    return (max(0,x))
```

```python
x = [float(x)/100 for x in range(-100,100)]   #Input
y = [relu(x1) for x1 in x]     #Output

plot_function(x,y,'ReLU Function')   # plot
```

### **Leaky ReLU**

```python
alpha = 0.1
def leaky_relu(x):
    if(x<0):
        return (alpha*x)
    else:
         return x
```

```python
x = [float(x)/100 for x in range(-100,100)]   #Input
y = [leaky_relu(x1) for x1 in x]     #Output

plot_function(x,y,'Leaky ReLU')   # plot
```

### **Sigmoid Function**

```python
from math import exp
def sigmoid(x):
    return (exp(x)/(1+exp(x)))
```

```python
x = [float(x)/100 for x in range(-500,500)]   #Input
y = [sigmoid(x1) for x1 in x]     #Output

plot_function(x,y,'Sigmoid Function')   # plot
```

### **Binary Function**

```python
def binary(x):
    if(x>0):
        return 1
    else:
        return 0
```

```python
x = [float(x)/100 for x in range(-500,500)]   #Input
y = [binary(x1) for x1 in x]     #Output

plot_function(x,y,'Binary Function')   # plot
```

### **Step Function**

```python
alpha =2
def step(x):
    if(x<0):
        return 0
    else:
        return alpha
```

```python
x = [float(x)/100 for x in range(-500,500)]   #Input
y = [step(x1) for x1 in x]     #Output

plot_function(x,y,'Step Function')   # plot
```

### **Linear Function**

```python
y_intercept=0
slop=1
def linear(x):
    return slop*x+y_intercept
```

```python
x = [float(x)/100 for x in range(-500,500)]   #Input
y = [linear(x1) for x1 in x]     #Output

plot_function(x,y,'Linear Function')   # plot
```

### **TanH Function**

```python
#tanh(x) = 2*sigmoid(2x) - 1  => 2/(1 + e-2x) - 1
def tanh(x):
    return 2*sigmoid(2*x)
```

```python
x = [float(x)/100 for x in range(-500,500)]   #Input
y = [tanh(x1) for x1 in x]     #Output

plot_function(x,y,'TanH Function')   # plot
```

### **Swish Function**

```python
beta =1
def swish(x):
    return 2*x*sigmoid(beta*x)
```

```python
x = [float(x)/100 for x in range(-500,500)]   #Input
y = [swish(x1) for x1 in x]     #Output

plot_function(x,y,'Swish Function')   # plot
```
